```python
data.info()
```

```python
data2 = data.copy()
data2 = data2.drop(['unnamed:_0'], axis = 1)
data2 = data2.reset_index(drop = True)
data2.describe().T
```

```python
data2
```

```python
data2.describe(include='all').T
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
response_counts = data2['response'].value_counts()
```

```python
response_counts.plot(kind='bar', stacked=True)
plt.xlabel('response')
plt.ylabel('Count')
plt.title('Total Counts of Response')
plt.show()
```

```python
crosstab = pd.crosstab(data2['response'], data2['sales_channel'])

crosstab.plot(kind='bar', stacked=True)
plt.xlabel('sales_channel')
plt.ylabel('Count')
plt.title('Counts of response by sales channel')
plt.show()
```

```python
sns.barplot(x="response", y="total_claim_amount", data=data2)
plt.title('Counts of response by total claim amount')
plt.show()
```

```python
sns.barplot(x="response", y="income", data=data2)
plt.title('Counts of response by income')
plt.show()
```
