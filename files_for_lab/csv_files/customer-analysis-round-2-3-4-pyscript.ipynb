```python
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
response_counts = data2['response'].value_counts()
```

```python
response_counts.plot(kind='bar', stacked=True)
plt.xlabel('response')
plt.ylabel('Count')
plt.title('Total Counts of Response')
plt.show()
```
